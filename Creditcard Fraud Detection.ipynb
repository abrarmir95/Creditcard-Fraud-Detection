```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('creditcard_fraud.csv')
```

```python
#Separting legit and fraudalent transactions respectively
df_legit=df[df['Class']==0]
df_fraud=df[df['Class']==1]
```

```python
df_legit.shape
```

```python
df_fraud.shape
```

```python
#Compare the values for both transactions

df.groupby("Class").mean()
```

```python
#Using under sampling taking equal no. of samples from both classes of dataset

df_legit_sample=df_legit.sample(n=492)
```

```python
df_legit_sample.shape
```

```python
df_fraud.shape
```

```python
#Concatenating two of above dataframes

df_new=pd.concat([df_legit_sample,df_fraud], axis=0)
```

```python
df_new.shape
```

```python
#Splitting the new dataset into features and target data

X=df_new.iloc[:,:-1]
y=df_new['Class']
```

```python
X
```

```python
y
```

```python
#Training test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,stratify=y, test_size=0.2, random_state=45)
```

```python
#Building Random forest classifier
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()

rf.fit(X_train,y_train)
```

```python
#Prediction

pred=rf.predict(X_test)
```

```python
#Evaluating and Visualizing model performance using confusion matrix
from sklearn.metrics import confusion_matrix
import scikitplot.plotters as skplt
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test, pred)*100
print(accuracy)
skplt.plot_confusion_matrix(y_test,pred)
```
